```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('../Datasets/ADS/total_training_data.csv')
train_df.head()
```

Looks like the column names are just numbered. Since we have the `feature_description.csv`, we can use that to add names to our variable columns

```python
feature_description = pd.read_csv('../Datasets/source/feature_description.csv', header=None)
feature_description.columns = ['Feature_Name','Feature_Detail']
feature_description
```

```python
train_df = train_df.rename(columns=dict(zip(train_df.columns[:88], feature_description['Feature_Name'])))
train_df.columns
```

```python
test_df = pd.read_csv('../Datasets/ADS/total_testing_data.csv')
test_df.head()
```

```python
test_df = test_df.rename(columns=dict(zip(test_df.columns[:88], feature_description['Feature_Name'])))
test_df.columns
```

```python
train_df
```

```python
#Save the files
train_df.to_csv('../Datasets/ADS/final_training_data.csv')
test_df.to_csv('../Datasets/ADS/final_testing_data.csv')
```
